```python
import pandas as pd

# Cargar archivos
try:
    cities = pd.read_csv("Cities.csv")
except FileNotFoundError:
    print("Error: 'Cities.csv' not found. Please ensure the file is in the correct directory.")
    cities = None # Assign None to avoid NameError later if not found

try:
    game_cities = pd.read_csv("MGameCities.csv")
except FileNotFoundError:
    print("Error: 'MGameCities.csv' not found. Please ensure the file is in the correct directory.")
    game_cities = None # Assign None to avoid NameError later if not found

# Mostrar tablas
if cities is not None:
    print(cities.head())
if game_cities is not None:
    print(game_cities.head())
```

```python
# Unir tablas (DESNORMALIZACIÓN)

tabla_desnormalizada = pd.merge(
    game_cities,
    cities,
    on='CityID',
    how='left'
)

# Mostrar resultado
print(tabla_desnormalizada.head())
```

```python
tabla_desnormalizada.to_excel('tabla_desnormalizada.xlsx', index=False)
```
